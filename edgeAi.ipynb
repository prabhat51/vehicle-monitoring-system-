```python
input_file = '/home/yuvraj/YuvrajML/vjproject/project/deep-text-recognition-benchmark/PaddleOCR_dataset/val_label.txt'
output_file = '/home/yuvraj/YuvrajML/vjproject/project/deep-text-recognition-benchmark/PaddleOCR_dataset/fixed_val_label.txt'

with open(input_file, "r") as f:
    lines = f.readlines()

with open(output_file, "w") as f:
    for line in lines:
        parts = line.strip().split('\t')
        if len(parts) == 2:
            img_path = parts[0].split('/')[-1]  # just filename
            f.write(f"{img_path}\t{parts[1]}\n")

```
